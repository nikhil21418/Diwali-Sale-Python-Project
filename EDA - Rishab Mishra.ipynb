```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')

```

```python
df = df.drop(columns=['Status', 'unnamed1'])
```

```python
df= df.dropna(subset=['Amount'])
```

# Exploratory Data Analyst

Gender

```python
ax = sns.countplot(x= 'Gender', data=df)
ax.bar_label(ax.containers[0])
plt.title('Gender')
plt.show()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)

ax = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
ax.bar_label(ax.containers[0])
```

Age

```python
df.columns
```

```python
ax= sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```

```python
ab = df.groupby(['Age Group'], as_index=False)['Orders'].sum().sort_values(by= 'Orders', ascending=False)
ax = sns.barplot(x = 'Age Group', y = 'Orders', data = ab)
ax.bar_label(ax.containers[0])
```

```python
ab = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
ax = sns.barplot(x = 'Age Group', y = 'Amount', data = ab)
```

state

```python
plt.figure(figsize=(12,8))
stat = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by= 'Orders', ascending=False).head(10)
ax = sns.barplot(x = 'State', y = 'Orders', data = stat, palette='Set1')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(12,8))
stat = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False).head(10)
ax = sns.barplot(x = 'State', y = 'Amount', data = stat, palette='Set1')
# for bars in ax.containers:
#     ax.bar_label(bars)
plt.xticks(rotation=45)
```
